## Описание проекта
### Задача
Подготовить прототип модели машинного обучения для компании «Цифры», которая разрабатывает решения для эффективной работы промышленных предприятий.

Модель должна предсказать коэффициент восстановления золота из золотосодержащей руды. 

### Данные:

    gold_industry_train.csv — обучающая выборка;
    gold_industry_test.csv — тестовая выборка;
    gold_industry_full.csv — исходные данные.

Данные индексируются датой и временем получения информации (признак date). Соседние по времени параметры часто похожи.

Исходный датасет содержит обучающую и тестовую выборки со всеми признаками.
### Описание данных
    Rougher feed — исходное сырье
    Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
      
    Xanthate — ксантогенат (промотер, или активатор флотации);
    Sulphate — сульфат (на данном производстве сульфид натрия);
    Depressant — депрессант (силикат натрия).
    Rougher process (англ. «грубый процесс») — флотация
    Rougher tails — отвальные хвосты
    Float banks — флотационная установка
    Cleaner process — очистка
    Rougher Au — черновой концентрат золота
    Final Au — финальный концентрат золота

Параметры этапов

    air amount — объём воздуха
    fluid levels — уровень жидкости
    feed size — размер гранул сырья
    feed rate — скорость подачи

Наименование признаков
Наименование признаков должно быть такое:
[этап].[тип_параметра].[название_параметра]
Пример: rougher.input.feed_ag
Возможные значения для блока [этап]:

    rougher — флотация
    primary_cleaner — первичная очистка
    secondary_cleaner — вторичная очистка
    final — финальные характеристики

Возможные значения для блока [тип_параметра]:

    input — параметры сырья
    output — параметры продукта
    state — параметры, характеризующие текущее состояние этапа
    calculation — расчётные характеристики
### Формулы 
Recovery:

$$ Recovery = \frac{C * (F-T) }{F*(C-T)}*100\% $$
sMape:

$$ sMape = \frac{1} {N} \sum_{i=1}^{N} \frac{|Y_i - y_i|}{(|Y_i|+|y_i|) / 2}*100\% $$

итоговое sMape:

$$ sMape = 25\% * sMape(rougher) + 75\% * sMape(final) $$

### Проверим расчеты обогащения

ииисмоделируем процесс восстановления золота из золотосодержащей руды.
Для этого расчитаем эффективность обогащения по формуле: $$ Recovery = \frac{C * (F-T) }{F*(C-T)}*100\% $$
где:
*   C — доля золота в концентрате после флотации/очистки;
*   F — доля золота в сырье/концентрате до флотации/очистки;
*   T — доля золота в отвальных хвостах после флотации/очистки.
 
 для признака ` rougher.output.recovery ` на обучающей выборке

```python
def recovery_func(c, f, t):
    return ((c * (f - t)) / (f * (c -t))) * 100
```

<font color='blue'><b>Комментарий ревьюера: </b></font> ✔️\
<font color='green'>Расчёт верный!</font>

функция для расчита эффективности обогащения руды

```python
tmp = mean_absolute_error(
    gold_industry['gold_industry_train'][
        [
            'rougher.output.concentrate_au',
            'rougher.input.feed_au', 
            'rougher.output.tail_au'
        ]
    ].agg(lambda x: recovery_func(*x), axis=1),
    gold_industry['gold_industry_train']['rougher.output.recovery'])
f"MAE для золота равна {tmp:.20f}"
```

учитывая, что MAE ничтожно мала, можно говорить о том, что расчеты в представленом наборе данных выполнены согласно формуле эффективности обогащения.

```python
def build_histplot(res_tree:list,
                   titles:str,
                   data_sets=gold_industry,
                   ds_key:str='gold_industry_full', 
                   one:bool=True,
                   a:int=20, b:int=5,
                   r_ax=False,
                   stat:str='count',
                   ylabe='количество',
                   xlabe='распределение',
                   lims=1500)->plt:
    #fig, ax = plt.subplots(ncols=len(res_tree), figsize=(a,b))
    fig, ax = plt.subplots(ncols=1, figsize=(a,b))
    color = sns.palettes.color_palette()
    for num, i in enumerate(res_tree):
            
    
        sns.histplot(
            data_sets[ds_key][i],
            color=color[num],
            #ax=ax if one else ax[num],
            ax=ax,
            legend=True,
            element="poly",
            stat=stat,
        );
    #plt.artist =res_tree
    plt.suptitle(titles)
    plt.legend(res_tree)
    ax.set_xlabel(xlabe)
    ax.set_ylabel(ylabe)
    if stat=='count':
        plt.ylim(top=lims)
    return ax
```

## функция для расчета sMAPE

```python
def s_mape(valid_data: pd.Series,
           predict: np.array)->float:
    valid_data = np.array(valid_data)
    numerator = np.abs(valid_data - predict)
    denominator = np.abs(valid_data) + np.abs(predict)
    result = round(np.mean(numerator / (denominator/2))*100, 2)
    return result
```

в функции производится расчет симметричного среднего абсолютного процентного отклонения по формуле: $$ sMape = \frac{1} {N} \sum_{i=1}^{N} \frac{|Y_i - y_i|}{(|Y_i|+|y_i|) / 2}*100\% $$
метрика выражается относительных величинах и одинаково учитывает масштаб и целевого признака, и предсказания.

```python
def result_s_mape(s_mape_rougher:float,
                  s_mape_final:float)->float:
    
    return s_mape_rougher*0.25 + s_mape_final*0.75
```

<font color='blue'><b>Комментарий ревьюера: </b></font> ✔️\
<font color='green'> `s_mape` и `result_s_mape` расчитаны верно.</font>

функция для расчета итогового sMape по формуле $$ sMape = 25\% * sMape(rougher) + 75\% * sMape(final) $$

```python
def model_func(model,
               train: pd.DataFrame,
               train_valid: pd.Series,
               param_grid:dict=dict(),
               random_state=None,
               name_data:str='s_mape'
              ):
    
    model_ridge = model(random_state=RANDOM_STATE) if random_state else model()

    gs_ridge = GridSearchCV(
        model_ridge, 
        param_grid=param_grid, 
        scoring=make_scorer(s_mape, greater_is_better=False),
        n_jobs=-1, 
        cv=5,
    )

    gs_ridge.fit(train, train_valid)
    best_model = gs_ridge.best_estimator_
    best_param = gs_ridge.best_params_
    data = {
        'name': [model.__name__],
        name_data: [abs(gs_ridge.best_score_)],
        'best_model': [f'{best_model=}'],
        'best_param': [f'{best_param=}'],
        'model': [f'model={model.__name__}'],
        #name_data: [tmp],
    }
    return gs_ridge, pd.DataFrame(data )
```

```python
def random_forest_regressor(train_, train_valid_, param=dict()):
    _, tmp = model_func(model=RandomForestRegressor,
                     param_grid={
                         'max_depth':np.arange(9, 12, 1),
                         'n_estimators':np.arange(100, 111, 5),
                         #'max_features':np.arange(3, 6, 1),
                         'min_samples_split':np.arange(2, 5, 1),
                         #'max_features': ['sqrt','log2'],
                         #'bootstrap': [True, False],
                     } if not param else param,
                     train=train_,
                     train_valid=train_valid_, 
                     random_state=RANDOM_STATE,)
    return tmp
```

```python
def model_ridge(train_, train_valid_, param=dict()):
    _, tmp = model_func(model=Ridge,
                           
                     param_grid={
                         'alpha': np.arange(0.0, 0.21, 0.01),
                         'solver':['auto', 'lsqr']
                     } if not param else param,
                     train=train_,
                     train_valid=train_valid_, 
                     random_state=RANDOM_STATE,)
    return tmp
```

```python
def dummy_regressor(train_, train_valid_, param=dict()):
    _, tmp = model_func(model=DummyRegressor,
                     
                     param_grid={'strategy': ['mean']} if not param else param,
                     train=train_,
                     train_valid=train_valid_,)
    return tmp
```

```python
def linear_regression(train_, train_valid_, param=dict()):
    _, tmp = model_func(model=LinearRegression,
                     param_grid={} if not param else param,
                     train=train_,
                     train_valid=train_valid_,)
    return tmp
```
